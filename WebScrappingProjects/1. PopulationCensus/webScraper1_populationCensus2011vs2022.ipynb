```python
import requests
import bs4
```

```python
#Requesting the page:
res= requests.get("https://www.findeasy.in/top-indian-states-by-population/")
```

```python
soup= bs4.BeautifulSoup(res.text,'lxml')
```

```python
type(soup)
```

```python
#Getting/Selecting the required table data:
soup.select("td")
```

### Above data not only consists of our required table but also some other tables with same "td" tag, so it requires some cleaning.

# Cleaning:

```python
#check: grabbing a text within the bs4.element.Tag:
soup.select("td")[0].text
```

```python
#grabbing the text of required table in a list:
lis=[]
rank=0
for items in soup.select("td"):
    if rank==2:
        break
    else:
        if items.text=="Rank":
            rank+=1
        lis.append(items.text)
        print(items.text)
print(lis)
```

```python
rank
```

```python
len(lis)
```

```python
#removing extra rank at last index:
lis.pop()
```

```python
#example try:
j='16.08\xa0 Lakh'.split("\xa0")
x=" ".join(j)
x
```

```python
#removing "\xa0" terms appearing in some places:
for x in range(0,len(lis)):
    if "\xa0" in lis[x]:
        print(x)
        j=lis[x].split("\xa0")
        lis[x]=" ".join(j)
```

```python
#new cleaned list:
lis
```

# Creating an array using list to be presented in more suitable(representable) form:

```python
arr=[]
count=0
while count!=len(lis)-4:
    arr.append([lis[count],lis[count+1],lis[count+2],lis[count+3]])
    count+=4
```
